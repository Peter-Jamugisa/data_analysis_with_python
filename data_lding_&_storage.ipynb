```python
#reading from an excel file

from openpyxl import load_workbook

# Load the Excel file
workbook = load_workbook(filename="C:/Users/HP/Downloads/sample.xlsx")

# Select the worksheet
worksheet = workbook['sheet1']

# Iterate over rows and print the cell values
for row in worksheet.iter_rows(min_row=1, values_only=True):
    for cell in row:
        print(cell)

# Close the workbook
workbook.close()

```

```python
#writing and reading from an excel file.
import pandas as pd


# Create a sample DataFrame
data = {'Name': ['John', 'Alice', 'Bob'],
        'Age': [25, 28, 30],
        'City': ['New York', 'Paris', 'London']}

frame = pd.DataFrame(data)

writer=pd.ExcelWriter("C:/Users/HP/Downloads/sample.xlsx", engine='openpyxl')
frame.to_excel(writer, "sheet1")

# writer.save()
writer.close()


from openpyxl import load_workbook

# Load the Excel file
workbook = load_workbook(filename="C:/Users/HP/Downloads/sample.xlsx")

# Select the worksheet
worksheet = workbook['sheet1']

# Iterate over rows and print the cell values
for row in worksheet.iter_rows(min_row=1, values_only=True):
    for cell in row:
        print(cell)

# Close the workbook
workbook.close()
```
